```python
with open('player_mvp_stats.csv', 'r', encoding='iso-8859-1') as f:
    csv_reader = csv.reader(f, delimiter=';')
    data = []
    for row in csv_reader:
        data.append(row)
df = pd.DataFrame(data[1:], columns=data[0])
df.head()
```

```python
df['Pos'] = df['Pos'].replace('PG', 1)
df['Pos'] = df['Pos'].replace('SG', 1)
df['Pos'] = df['Pos'].replace('PG-SG', 1)
df['Pos'] = df['Pos'].replace('SF-PG', 2)
df['Pos'] = df['Pos'].replace('PG-SF', 1)
df['Pos'] = df['Pos'].replace('PG-PF', 1)
df['Pos'] = df['Pos'].replace('SG-PG', 1)
df['Pos'] = df['Pos'].replace('SG-SF', 1)
df['Pos'] = df['Pos'].replace('SF-SG', 2)
df['Pos'] = df['Pos'].replace('SG-PF', 1)
df['Pos'] = df['Pos'].replace('PF-SG', 2)
df['Pos'] = df['Pos'].replace('SF', 2)
df['Pos'] = df['Pos'].replace('PF', 2)
df['Pos'] = df['Pos'].replace('SF-PF', 2)
df['Pos'] = df['Pos'].replace('PF-SF', 2)
df['Pos'] = df['Pos'].replace('PF-C', 2)
df['Pos'] = df['Pos'].replace('C-PF', 3)
df['Pos'] = df['Pos'].replace('SF-C', 2)
df['Pos'] = df['Pos'].replace('SF-C', 2)
df['Pos'] = df['Pos'].replace('C-SF', 3)
df['Pos'] = df['Pos'].replace('C', 3)
```

```python
df[['Pos','PTS', 'STL', 'AST', 'BLK', 'FG%', '3P%', '2P%','TOV', 'W/L%', 'FGA', '3PA', '2PA']] = df[['Pos','PTS', 'STL', 'AST', 'BLK', 'FG%', '3P%','2P%','TOV', 'W/L%', 'FGA', '3PA', '2PA']].apply(pd.to_numeric)
df_avg = df.groupby(['Player'])[['Pos','PTS', 'STL', 'AST', 'BLK', 'FG%', '3P%','2P%','TOV', 'W/L%', 'FGA', '3PA', '2PA']].mean()
df_avg = df_avg.reset_index()
df_finalavgs = df_avg.dropna()
df_finalavgs.head(50)
```
